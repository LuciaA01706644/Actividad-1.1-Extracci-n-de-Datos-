```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Filtro por objeto 
#Las variable filtrada fue el nombre de un cliente, fue identificado por medio de la columna "nombre_cliente"
filtro1=df[df["NOMBRE_CLIENTE"] == "SEVEN DAYS ALL SPORT"]
filtro1
```

```python
#Filtro por filas
#Las variables filtradas fueron los datos de las filas 100 a la 149
filtro2= df.iloc[100:150,: ]           #Filas consecutivas
#filtro2=df.iloc[ [10, 24, 100] , : ]  #Filas no consecutivas 
filtro2
```

```python
#Filtro por columnas
#Las variables filtradas fueron las columnas de "clave del articulo", "Costo total calculado" y "margen total calculado"

filtro3=df.iloc[ : , [5,11,13]]  #Columnas no consecutivas 
filtro3
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
#Se asignó la columna "Bombre_vendedor" como indice y se filtró con el nombre de los vendedores "Leticia Ramirez" y "Diego Armando Leon" con el "Margen Total Calulado"
filtro4=df1.loc[["LETICIA RAMIREZ HERNANDEZ","DIEGO ARMANDO LEON DURAN"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
#Filtro por cabecera
#Se filtró para que solamente se muestren las primeras 10 filas de datos
filtro5=df.head(10)
filtro5
```

```python
#Filtro por comparacion
#Se filtró para que  muestre los datos de los artículos y su compra que tengan un margen total calculado mayor a $20000
filtro6=df[df["MARGEN_TOTAL_CALCULADO"] > 20000]
filtro6
```

```python
#Filtro Y
#Se filtró para que muestre los datos de los artículos comprados que tengan un margen total calculado mayor a $20000 Y Hayan sido vendidos por Leticia Ramirez
filtro7=df[(df["MARGEN_TOTAL_CALCULADO"] > 20000) & (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro7
```

```python
#Se filtraron las variables de los datos que tengan un margen total calculado mayor a $20,000 y su factura haya tenido una fecha del 22/07/2022
filtro8=df[(df["MARGEN_TOTAL_CALCULADO"] > 20000) | (df["FECHA_DOC"] == "2022-07-22")]
filtro8
```

```python
#Filtro not
#Se filtraron las variables que no tengan un marggen total calculado mayor a 20000 y no tengan una fecha de emisión del 22/07/2022
filtro9=df[~(df["MARGEN_TOTAL_CALCULADO"] > 20000) & ~(df["FECHA_DOC"] == "2022-07-22")]
filtro9
```
